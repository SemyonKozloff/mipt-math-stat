```python
import numpy as np
import scipy.stats as sps
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
N = 10000
A = 1
B = 10

class cont_distr_gen(sps.rv_continuous):
    def _pdf(self, x):
        return 5 / (x ** 6)

cont_distr = cont_distr_gen(a=A, b=B, name='cont_distr')
```

```python
sample = cont_distr.rvs(size=N)
grid = np.linspace(A, B, 1000)
#mean = cont_distr.mean()
var = cont_distr.var() 

plt.figure(figsize=(16, 10))
plt.xlim(1.2, 5)
plt.ylim(-0.2, 2)
plt.scatter(sample, np.zeros(N), alpha=0.2, label='sample')
plt.plot(grid, cont_distr.pdf(grid), color='red', label='density')
plt.grid()
plt.title(r'$p_X(x)=5x^{-6}$', fontsize=16)
plt.legend(fontsize=16)
plt.show()
#plt.savefig("cont_distr.pdf")

var = cont_distr.var() 
delta_range = np.zeros(N)
for n in range(1, N):
    s_var = sample[:n].var()
    delta_range[n] = abs(var - s_var)
    
plt.figure(figsize=(16, 10))
plt.ylim(0, 0.06)
plt.plot(range(1, N), delta_range[1:N], color='green', label=r'$|S^{2}[X] - D[X]|$')
plt.grid()
plt.legend(fontsize=16)
plt.show()
#plt.savefig("delta_cont_distr.pdf")
```

```python
sample = sps.cauchy.rvs(size=N)
grid = np.linspace(-20, 20, 1000)

plt.figure(figsize=(16, 10))
plt.xlim(-15, 15)
plt.scatter(sample, np.zeros(N), alpha=0.2, label='sample')
plt.plot(grid, sps.cauchy.pdf(grid), color='red', label='density')
plt.grid()
plt.title(r'$p_{C(0, 1)}(x)$', fontsize=16)
plt.legend()
plt.show()
#plt.savefig("cauch_distr.pdf")

delta_range = np.zeros(N)
for n in range(1, N):
    s_var = sample[:n].var()
    delta_range[n] = s_var
    
plt.figure(figsize=(16, 10))
plt.plot(range(1, N), delta_range[1:N], color='green', label=r'$S^{2}[X]$')
plt.legend(fontsize=16)
plt.grid()
plt.show()
#plt.savefig("delta_cauchy.pdf")
```
