```python
import numpy as np
import scipy.stats as sps
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
N = 10000
THETA = 1

def plot_estimation(estimate, color, label):
    delta_range = np.zeros(N)   
    for n in range(1, N):
        estimation = estimate(sample[:n])
        delta_range[n] = abs(THETA - estimation)
    
    plt.plot(range(1, N), delta_range[1:N], color=color, label=label)
    
    
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'olivedrab', 'grey', 'purple']
    
```

```python
plt.figure(figsize=(16, 10))
plt.ylim(0, 2.5)

color_id = 0
sample = sps.expon.rvs(loc=0, scale=THETA, size=N)

for k in range(5, 50, 5):
    estimate = lambda s : (np.math.factorial(k) / np.power(s, k).mean()) ** (1 / k)
    plot_estimation(estimate, colors[color_id], r'$k=' + str(k) + r'$')
    color_id = (color_id + 1) % len(colors)
    
plt.grid()
plt.title(r'$\theta=' + str(THETA) + r'$', fontsize=16)
plt.legend(fontsize=16)
plt.show()
#plt.savefig("k-5-50-5.pdf")
```
