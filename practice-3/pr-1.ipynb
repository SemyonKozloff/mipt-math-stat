```python
import numpy as np
import scipy.stats as sps
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
M = 100
N = 1000

theta_values = [7, 15, 31]
square_func = [[]] * M

def plot_function(estimate, color, label): 
    for m in range(M):
        sample = sps.uniform.rvs(loc=0, scale=theta, size=N)
        estimation = np.array([estimate(sample[:n]) for n in range(1, N + 1)])
        square_func[m] = (estimation - theta) ** 2
        
    average_square_func = [np.mean([square_func[m][n] for m in range(M)])
                               for n in range(1, N)] 
    plt.plot(range(1, N), average_square_func, color=color, label=label)
```

```python
for theta in theta_values:
    plt.figure(figsize=(16,10))
    plt.ylim(-0.1, theta / 7)
    
    plot_function(lambda s : 2 * s.mean(), 'red', r'$2\overline{X}$')
    plot_function(lambda s : s.mean() + s.max() / 2, 'green', r'$\overline{X}+X_{(n)}/2$')
    #plot_function(lambda s : (s.size + 1) * s.min(), 'yellow', r'$(n + 1)X_{(1)}$')
    #plot_function(lambda s : s.max() + s.min(), 'blue', r'$X_{(1)}+X_{(n)}$')
    #plot_function(lambda s : (s.size + 1) / s.size * s.max(), 'purple', r'$\frac{n+1}{n}X_{(n)}$')
       
    plt.grid()
    plt.title(r'$\theta=' + str(theta) + r'$', fontsize=16)
    plt.legend(fontsize=16)
    plt.show()
    
```

```python
for theta in theta_values:
    plt.figure(figsize=(16,10))

    plot_function(lambda s : (s.size + 1) * s.min(), 'yellow', r'$(n + 1)X_{(1)}$')
       
    plt.grid()
    plt.title(r'$\theta=' + str(theta) + r'$', fontsize=16)
    plt.legend(fontsize=16)
    plt.show()
    
```

```python
for theta in theta_values:
    plt.figure(figsize=(16,10))
    plt.ylim(-0.1, theta / 40)
    
    plot_function(lambda s : s.max() + s.min(), 'blue', r'$X_{(1)}+X_{(n)}$')
    plot_function(lambda s : (s.size + 1) / s.size * s.max(), 'purple', r'$\frac{n+1}{n}X_{(n)}$')
       
    plt.grid()
    plt.title(r'$\theta=' + str(theta) + r'$', fontsize=16)
    plt.legend(fontsize=16)
    plt.show()
    
```
