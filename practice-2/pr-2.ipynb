```python
import numpy as np
import scipy.stats as sps
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def plot_likelihood(grid, sample, f):
    likelihood = [f(sample, x) for x in grid]
    left_border = grid[0]
    step = round(grid[1] - grid[0], 3)
    argmax = left_border + np.argmax(likelihood) * step
    
    plt.plot(grid, likelihood, color='red')
    
    return argmax
    
```

```python
with open('Cauchy.txt', 'r') as input_file:
    sample = list(map(float, input_file.readlines()))
N = len(sample)
grid = np.arange(-1000, 1000, 0.01)

f = lambda s, t : -np.sum(np.log(1 + (s - t) ** 2))
plt.figure(figsize=(16, 10))
argmax = plot_likelihood(grid, sample, f)
plt.xlabel(r'$argmax{f_{\theta}(X_1,...,X_N)}=' + str(argmax) + r'$', fontsize=16)
plt.grid()
plt.title(r'$N=' + str(N) + r'$', fontsize=16)
plt.show()

N = int(N / 2)
sample = sample[:N]
plt.figure(figsize=(16, 10))
argmax = plot_likelihood(grid, sample, f)
plt.xlabel(r'$argmax{f_{\theta}(X_1,...,X_N)}=' + str(argmax) + r'$', fontsize=16)
plt.grid()
plt.title(r'$N=' + str(N) + r'$', fontsize=16)
plt.show()
```

```python
with open('Weibull.txt', 'r') as input_file:
    sample = list(map(float, input_file.readlines()))
N = len(sample)
grid = np.arange(-2, 2, 0.001)

f = lambda s, t : t * np.log(10) + (np.exp(t * np.log(10)) - 1) * np.sum(np.log(s))- np.sum(s ** np.exp(t * np.log(10)))
plt.figure(figsize=(16, 10))
argmax = plot_likelihood(grid, sample, f)
plt.xlabel(r'$argmax(ln(f_{\gamma}(X_1,...,X_N)))=10^{' + str(argmax) + r'}$', fontsize=16)
plt.grid()
plt.title(r'$N=' + str(N) + r'$', fontsize=16)
plt.show()

N = int(N * 0.4)
sample = sample[:N]
plt.figure(figsize=(16, 10))
argmax = plot_likelihood(grid, sample, f)
plt.xlabel(r'$argmax(ln(f_{\gamma}(X_1,...,X_N)))=10^{' + str(argmax) + r'}$', fontsize=16)
plt.grid()
plt.title(r'$N=' + str(N) + r'$', fontsize=16)
plt.show()
```
